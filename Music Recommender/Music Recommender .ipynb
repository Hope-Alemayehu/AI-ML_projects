We have an online music store and when our users sign up.we ask their age and gender and based on their profile we recommend various music albums they are likely to buy.

In this project, we want to use machine learning to increase sales

The list of steps in ML projects:
    1. import data
    2.clean data
    3.split the data
    4.choose the model
    5.train the model
    6.test the model
    7.evaluate and improve

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib


music_data = pd.read_csv("music.csv")
```

```python
#we are spliting the dataset to create other two datasets
X = music_data.drop(columns=["genre"])
Y = music_data["genre"]
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
model = DecisionTreeClassifier()
#we need to train it so it learns
model.fit(x_train, y_train)
prediction = model.predict(x_test)
score = accuracy_score(y_test, prediction)
score
```
